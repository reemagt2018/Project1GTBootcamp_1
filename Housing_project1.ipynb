```python
# 1. What are the geocoordinates (latitude and longitude) of Seattle,
# Washington?
target_city = input("Please input the city where you would like to move : ")
distance=int(input("Please input the radius around the city : "))

params = {"address": target_city, "key": gkey}

# Build URL using the Google Maps API
base_url = "https://maps.googleapis.com/maps/api/geocode/json"

print(f" The Geocoordinates of {target_city} is")

# Run request
response = requests.get(base_url, params=params)
print(response)

# Convert to JSON
target_geo = response.json()
pprint.pprint(target_geo)

# Extract lat/lng
lat = target_geo["results"][0]["geometry"]["location"]["lat"]
lng = target_geo["results"][0]["geometry"]["location"]["lng"]
state=target_geo["results"][0]["address_components"][2]["short_name"]

# Print results
print(f"{target_city}: {lat}, {lng}, {state}")
```
